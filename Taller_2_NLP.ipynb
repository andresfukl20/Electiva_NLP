```python
import urllib.request
import bs4 as bs
```

```python
link='https://es.wikipedia.org/wiki/Carl_Sagan'
```

```python
request=urllib.request.urlopen(link)
fuente=request.read()
request.close()
```

```python
#procesar la pagina web como HTML
soup=bs.BeautifulSoup(fuente,'html.parser')
soup.find('title')
#soup
```

```python
#trae el nombre de titulo
soup.find('title').contents
```

```python
soup.find(id="firstHeading").contents 
```

```python
for respuesta in soup.find('p').find_all('a'):
    print(respuesta)
    print('Contenido:')
    print(respuesta.contents)
    print('Título:')
    print(respuesta.get('title'))
    print()
```

```python
mi_lista=['https://es.wikipedia.org/wiki/Albert_Einstein','https://es.wikipedia.org/wiki/Carl_Sagan',
          'https://es.wikipedia.org/wiki/Carl_Friedrich_Gauss','https://es.wikipedia.org/wiki/Isaac_Newton',
         'https://es.wikipedia.org/wiki/Galileo_Galilei','https://es.wikipedia.org/wiki/Stephen_Hawking',
         'https://es.wikipedia.org/wiki/Charles_Darwin','https://es.wikipedia.org/wiki/Leonardo_da_Vinci',
         'https://es.wikipedia.org/wiki/Nikola_Tesla','https://es.wikipedia.org/wiki/Marie_Curie']
for pagina in mi_lista:
        request=urllib.request.urlopen(pagina)
        fuente=request.read()
        request.close()
        soup=bs.BeautifulSoup(fuente,'html.parser')
        print(soup.find(id="firstHeading").contents)
       
```

```python
for pag in mi_lista:
    for respuesta in soup.find('p').find_all('a'):
        request=urllib.request.urlopen(pag)
        fuente=request.read()
        request.close()
        soup=bs.BeautifulSoup(fuente,'html.parser')
        print(respuesta)
        print('Contenido:')
        print(respuesta.contents)
        print('Título:')
        print(respuesta.get('title'))
        print()
```
